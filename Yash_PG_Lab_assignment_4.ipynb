#**Data Exploration**

###Loading the iris dataset

```python
#Loading the iris dataset
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
iris_data = load_iris()
```

```python
iris_df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
iris_df['species'] = iris_data.target
```

```python
iris_df
```

###Plotting the histograms of Iris Features

```python
features = iris_df.columns[:-1]

# Plot histograms
iris_df[features].hist(bins=15, figsize=(10, 6))
plt.suptitle('Histograms of Iris Features')
plt.show()
```

###Plotting scatter plots of Iris features

```python
import seaborn as sns

# Pairplot to show scatter plots between each pair of features
sns.pairplot(iris_df, hue='species')
plt.suptitle('Scatter Plots of Iris Features')
plt.show()
```

###Calculating mean,median and standard deviation of Iris features

```python
#avarage of each feature
petal_lenght_average = iris_df['petal length (cm)'].mean()
sepal_length_average = iris_df['sepal length (cm)'].mean()
petal_width_average = iris_df['petal width (cm)'].mean()
sepal_width_average = iris_df['sepal width (cm)'].mean()

print("average petal lenght:",petal_lenght_average,"cm")
print("average sepal length:",sepal_length_average,"cm")
print("average petal width:",petal_width_average,"cm")
print("average sepal width:",sepal_width_average,"cm")
```

```python
#median of each feature
petal_lenght_median = iris_df['petal length (cm)'].median()
sepal_length_median = iris_df['sepal length (cm)'].median()
petal_width_median = iris_df['petal width (cm)'].median()
sepal_width_median = iris_df['sepal width (cm)'].median()

print("petal lenght median:",petal_lenght_median,"cm")
print("sepal length median:",sepal_length_median,"cm")
print("petal width median:",petal_width_median,"cm")
print("sepal width median:",sepal_width_median,"cm")
```

```python
#standard deviation of each feature
petal_lenght_std = iris_df['petal length (cm)'].std()
sepal_length_std = iris_df['sepal length (cm)'].std()
petal_width_std = iris_df['petal width (cm)'].std()
sepal_width_std = iris_df['sepal width (cm)'].std()

print("petal lenght standard deviation:",petal_lenght_std)
print("sepal length standard deviation:",sepal_length_std)
print("petal width standard deviation:",petal_width_std)
print("sepal width standard deviation:",sepal_width_std)
```

#**Data Preprocessing**

###normalising the data

```python
X = iris_data.data

# Z-Score normalization((X-mue)/sigma) of data except labels
normalised_X = (X - X.mean()) / X.std()
```

```python
normalised_X
```

###Converting labels to one hot encoding

```python
from sklearn.preprocessing import OneHotEncoder

y = iris_data.target.reshape(-1,1)
y = OneHotEncoder(sparse=False).fit_transform(y)
```

#**Building the Neural Network**

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Define the model
model = Sequential()

# Add layers to the model
model.add(Dense(10, input_shape=(4,), activation='relu'))  # Input layer + first hidden layer
model.add(Dense(3, activation='softmax'))  # Output layer for 3 classes (Iris setosa, versicolor, virginica)
model.summary()
```

#**Training the Model**

```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(normalised_X, y, test_size=0.2, random_state=42)

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=40, batch_size=5, validation_split=0.2)
```

###Plot training & validation accuracy values

```python

plt.figure(figsize=(15, 6))

# Accuracy Plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='upper left')

# Loss Plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')

# Show the plots
plt.show()
```

#**Model Evaluation**

###Testing trained model on test data

```python
# Evaluate the model on the test set
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Loss: {loss}")
print(f"Test Accuracy: {accuracy}")
```

###Plotting the confusion matrix on test data

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

predictions = model.predict(X_test)
predictions = predictions.argmax(axis=1)

y_true = y_test.argmax(axis=1)

cm = confusion_matrix(y_true, predictions)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```
